A dataset on the job outcomes of students who graduated between 2010 and 2012 was utilized for this projc. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducted serveys and aggregated the data. FiveThirtyEight cleaned the dataset and released it on [Github repo](https://www.census.gov/programs-surveys/acs/). 

The aim of this project was to see how using pandas plotting functionality along with Jupyter notebook interface allowing us to explore data quicly using visualizations. Let's get started!

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0] #return the first row formatted as a table
```

```python
print(recent_grads.head())
print(recent_grads.tail())
```

```python
#generate summary statistics for all the numeric columns
recent_grads.describe()
```

```python
#drop rows containing missing values 
recent_grads = recent_grads.dropna()
```

## Pandas, Scatter Plots


```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter') 
```


## Pandas, Histograms

```python
cols=['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate','Men','Women']

fig=plt.figure(figsize=(5,12))
for r in range (1,5):
    ax=fig.add_subplot(4,1,r)
    ax=recent_grads[cols[r]].plot(kind='hist', rot=30)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig=plt.figure(figsize=(5,12))
for r in range (4,8):
    ax=fig.add_subplot(4,1,r-3)
    ax=recent_grads[cols[r]].plot(kind='hist', rot=40)
```

## Pandas, Scatter Matrix Plot

A scatter matrix plot would combine both scatter plots and histograms into 1 grid of plots and allows us to explore potential relationships and distributions simultaneoy.

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']],figsize=(6,6))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

## Pandas, Bar Plots

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', legend=False)
```
